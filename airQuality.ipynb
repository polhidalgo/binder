```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("airquality.csv")
```

```python
print(df.head(10))
```

```python
print("Nombre de registres:", len(df))
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
temp_max = df["Temp"].max()
print("Temperatura max:", temp_max)
```

```python
max_temp_row = df[df["Temp"] == temp_max]
print("Mes u dia amb temp maxima:")
print(max_temp_row[["Month", "Day"]])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Wind"], df["Solar.R"], alpha=0.7)
plt.xlabel("Wind")
plt.ylabel("Solar.R")
plt.show()
```

```python
## Permet visualitzar si hi ha alguna relacio entre vent i la radiacio solar.
```

```python
tempMonth = df.groupby("Month")["Temp"].mean()
```

```python
tempMonth.plot(kind="bar", figsize=(8,6))
```

```python
## Permet comparar la temperatura mitjana de cada mes
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["Ozone"], df["Temp"], alpha=0.7, color='green')
plt.xlabel("Ozo")
plt.ylabel("Temp")
plt.show()
```
